```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_rat(percent_hit=0.9, max_hit=10):
    # percent_hit: 1回あたりの攻撃が当たる確率。普通は0.9、広角レンズ持ちは0.99
    # max_hit: さ
    E=0
    P=1
    for i in range(max_hit+1):
        if i==max_hit:
            # 最後までヒットしたときのエスケープ
            E+=P*max_hit
        else:
            # ハズレの場合はi回のヒット
            E+=i*P*(1-percent_hit)
            P*=percent_hit
    return (E)
```

```python
print(calc_rat(0.5, 10))
```

```python
x=np.arange(0.8,1.01,0.01)
y=[calc_rat(i) for i in x]
```

```python
plt.plot(x, y, marker=".")

plt.xlabel("percent_hit")
plt.ylabel("E_hit")
```
